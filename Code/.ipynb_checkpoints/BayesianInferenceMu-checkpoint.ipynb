Bayesian update of mu for known sigma

\begin{equation} E(\mu' | x) = \frac{\sigma^2 \mu + \sigma_0^2 x}{\sigma^2 + \sigma_0^2} \end{equation}

\begin{equation} Var(\mu' | x) = \frac{\sigma \sigma_0}{\sigma^2 + \sigma_0^2} \end{equation}

```python
n = 100
mu_true = 10
sigma_true = .1
xa = np.random.normal(mu_true, sigma_true, n)
mu = xa
sigmaa = xa

# Prior
mu_0 = 5.0
sigma_0 = 5

mu[0] = mu_0
sigmaa[0] = sigma_0
for i in range(1,n):
    sigma = sigma_true
    x = xa[i]
    sigma_1 = np.sqrt((sigma**2*sigma_0**2)/(sigma**2+sigma_0**2))
    sigmaa[i] = sigma_1
    mu_1 = (mu_0*sigma**2+x*sigma_0**2)/(sigma**2+sigma_0**2)
    mu[i] = mu_1
sns.set()
plt.plot(mu)
```
